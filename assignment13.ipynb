```python
# Calculate the accuracy of the model 
from sklearn.metrics import accuracy_score

# Calculate the precision of the model
from sklearn.metrics import precision_score

# Calculate the recall of the model
from sklearn.metrics import recall_score

from sklearn.metrics import f1_score


```

```python
#cross su X_train 
from sklearn.metrics import accuracy_score, f1_score,classification_report

from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV

def ricerca(model, param):
  k = StratifiedKFold(n_splits=5, shuffle = True)
  gridsearch = GridSearchCV(model, param, scoring='accuracy', cv=k)
  gridsearch.fit(X_train, y_train)
  pred = gridsearch.predict(X_test)
  
  accuracy = accuracy_score(y_test, pred)
  precision = precision_score(y_test, pred, average='micro')
  recall = recall_score(y_test, pred, average='micro')

  cr = classification_report(y_test, pred)
  performances = [accuracy, precision, recall]
  return gridsearch.best_estimator_, gridsearch.best_score_, gridsearch, performances, cr

```

```python
#Ram not enough to try more combinations
performances  = []
param1 ={"n_estimators":[100]}
r = ricerca(RandomForestClassifier(), param1)
performances.append(r[3])
```

```python

param2 ={"criterion":["gini", "log_loss"]}
r = ricerca(DecisionTreeClassifier(), param2)
performances.append(r[3])
```

```python

param3 ={"criterion":["gini", "log_loss"], "splitter":["random", "best"]}
r = ricerca(ExtraTreeClassifier(random_state=0), param3)
performances.append(r[3])
```

```python

param4 ={"n_estimators":[5,10,15]}
r = ricerca(BaggingClassifier(extra_tree, random_state=0), param4)
performances.append(r[3])
```

```python


param5 ={"alpha":[0.5, 1.5,2,5]}
r = ricerca(RidgeClassifier(), param5)
performances.append(r[3])
```
